```python
import numpy as np
from datascience import *
from bs4 import BeautifulSoup 
import requests
import csv
%matplotlib inline
import matplotlib.pyplot as plots
#https://www.dataquest.io/blog/web-scraping-tutorial-python/ for beautiful soup

"""a function to get acess to an artist's page and lyrics using metrolyrics"""
def request_access(link):
    page = requests.get(link)
    #page
    soup = BeautifulSoup(page.content, 'html.parser')
    #list(soup.children)
    lyrics_description = soup.find_all(class_= "title hasvidtable")
    return lyrics_description

def make_table(lyrics_description):
    import re
    total_lyrics = Table(['title', 'lyrics'])
    for lyrics in lyrics_description:
        s = str(lyrics)
        href = s.find("href")
        end = s.find("onmousedown")
        link = s[href+6: end-2]
        name = s.find("title=")+6
        #getting name
        title = s[name:]
        end_name = s.find("lyrics")
        title = title[:end_name]
        #gets the actual lyrics from the title
        page = requests.get(link)
        page
        soup = BeautifulSoup(page.content, 'html.parser')
        soup.prettify()
        verses= soup.find_all(class_= "verse")
        temp = str(verses)
        get_rid_of = ['</br>', "<br>", '<p class="verse">', '</p>']
        for term in get_rid_of:
            temp = re.sub(term, '', temp)
            #print(temp)
        total_lyrics = total_lyrics.with_row([title, temp])
        #total_lyrics.show(5)
    return total_lyrics
        
        
        
"""a function to make a table to put the artist and lyrics together"""
def making_table(artist_name, total_lyrics):
    artist = make_array()
    for i in np.arange(total_lyrics.num_rows):
        artist = np.append(artist, artist_name)
    total_table_lyrics = total_lyrics.with_column("Artist", artist)
    #make csv file
    csv_name = str(artist_name) + ".csv"
    #print(csv_name)
    with open(csv_name, "w", newline = '', encoding='utf-8') as csvfile:
        names = ["title", "lyrics", "artist"]
        writer = csv.DictWriter(csvfile, fieldnames = names)
        writer.writeheader()
        for row in np.arange(total_lyrics.num_rows):
            writer.writerow({'title': total_table_lyrics.row(row)[0], 'lyrics': total_table_lyrics.row(row)[1], 'artist' :total_table_lyrics.row(row)[2]})
```

```python
lyrics_description = request_access('http://www.metrolyrics.com/kendrick-lamar-lyrics.html')
lyrics_description 
```

```python
kendrick_lyrics = make_table(lyrics_description)
```
